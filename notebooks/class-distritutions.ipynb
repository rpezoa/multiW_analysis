```python
import uproot
from os import path
from glob import glob
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data_path = "../../raw_data/trees/"
all_files = glob(f"{data_path}/*.root")
l=[]
total_back = 0
total_sig = 0
for file_s in all_files:
    #print(file_s)
    file = uproot.open(file_s)
    keys = file.keys()
    tree = file["T"]
    df =tree.pandas.df(tree.keys())
    f = file_s.split("/")[-1].split(".")[0]
    nrows,ncols = df.shape
    if "signal" in f:
        total_sig+=nrows
    else:
        total_back+=nrows
        
    d = {"sample": f, "n":nrows }
    l.append(d)
    #display(df.head())
    print(df.shape)
    
df_distro = pd.DataFrame(l)
```
