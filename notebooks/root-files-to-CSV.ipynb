```python
import ROOT
import pandas as pd
from root_numpy import tree2array
from os import path
from glob import glob
import seaborn as sns
from matplotlib import pyplot as plt
import uproot
```

```python
data_path = "../../raw_data/trees/"
proc_data_path = "../processed_data/"
signal = "signal_"
```

```python
def root2csv(root_file,signal=True, weight=None, sample_name=""):
    
    file = uproot.open(root_file)
    
    keys = file.keys()
    print(keys)
    tree = file["T"]
    df =tree.pandas.df(tree.keys())
    display(df.head())
    if signal:
        df["label"]=1
    else:
        df["label"]=0

    
    if weight != None:
            df["weight_column"] = weight
            
    df["sample"] = sample_name
            
    file_name = root_file.split("/")[-1].split(".")[0]
    df.to_csv(f"{proc_data_path}/{file_name}.csv", index = False)
    return df
```

```python
all_roots = glob(f"{data_path}/*.root")
print(all_roots)

signal_roots = glob(f"{data_path}/{signal}*.root")

bg_roots = list(set(all_roots) - set(signal_roots) ) 
```

```python
l_w = []
for sig_df in signal_roots:
    f = sig_df.split("/")[-1]
    print(f)
    print(sig_df)
    w = xsec_fb[f]*3000/n_events[f]
    l_w.append({"name": f,"weight":w})
    df = root2csv(sig_df,signal=True, weight=w, sample_name = sample_names[f])
```

```python
for bg_df in bg_roots:
    bg_name = bg_df.split("/")[-1]
    print(xsec_fb[bg_name])
    print(bg_df)
    w = xsec_fb[bg_name]*3000/n_events[bg_name]
    l_w.append({"name": bg_name,"weight":w})
    df = root2csv(bg_df,signal=False,weight=w, sample_name=sample_names[bg_name])
```
